```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df_num = pd.read_csv('./df_num.csv')
```

```python
sns.set_style('white',rc={'figure.figsize':(12,8)})
count_classes = pd.value_counts(df_num['PotentialFraud'], sort = True)
print("Percent Distribution of Potential Fraud class:- \n",count_classes*100/len(df_num))
LABELS = ["Non Fraud", "Fraud"]
#Drawing a barplot
count_classes.plot(kind = 'bar', rot=0,figsize=(10,6))

#Giving titles and labels to the plot
plt.title("Potential Fraud distribution in Aggregated claim transactional data")
plt.xticks(range(2), LABELS)
plt.xlabel("Potential Fraud Class ")
plt.ylabel("Number of PotentialFraud per Class ")

plt.savefig('./pics/PotentialFraudDistributionInMergedData')
```

```python
fig1, ax1 = plt.subplots()
sns.set(font_scale=2)
labels =["Non Fraud", "Fraud"]
ax1.pie(count_classes, labels=labels, autopct='%1.1f%%', shadow=True)
ax1.axis('equal')
plt.show()
plt.savefig('./pics/PotentialFraud_pie')
```

```python
df_provider = pd.read_csv('./train.csv')
```

```python
sns.set_style('white',rc={'figure.figsize':(12,8)})
count_classes_provider = pd.value_counts(df_provider['PotentialFraud'], sort = True)
print("Percent Distribution of Potential Fraud class:- \n",count_classes*100/len(df_provider))
LABELS = ["Non Fraud", "Fraud"]
#Drawing a barplot
count_classes.plot(kind = 'bar', rot=0,figsize=(10,6))

#Giving titles and labels to the plot

plt.xticks(range(2), LABELS)
plt.xlabel("Potential Fraud Class ")
plt.ylabel("Number of providers ")

plt.savefig('./pics/PotentialFraud_provider')
```

```python
fig1, ax1 = plt.subplots()
labels =["Non Fraud", "Fraud"]
ax1.pie(count_classes_provider, labels=labels, autopct='%1.1f%%', shadow=True)
ax1.axis('equal')

plt.savefig('./pics/PotentialFraud_merge_pie')
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
colors = sns.color_palette('pastel')
explode1 = [0,  0.05]
explode2 = [0,  0.18]
fig, axs = plt.subplots(1, 2)
sns.set(font_scale=1.5)

axs[0].pie(count_classes, labels=labels, autopct='%1.1f%%', shadow=True,explode = explode1,colors = colors)

axs[1].pie(count_classes_provider, labels=labels, autopct='%1.1f%%', shadow=True,explode = explode2,colors = colors)

axs[0].set_title('Merged',fontsize=30)
axs[1].set_title('Original',fontsize=30)



plt.savefig('./pics/PotentialFraud_2pies')


```

```python
res=[]
for i in range(1000,38000,1000):
    x=df_num[(df_num['InscClaimAmtReimbursed']>=(i-1000)) & 
             (df_num['InscClaimAmtReimbursed']<=i) ]['PotentialFraud'].value_counts('yes')[0]
    res.append([i,x])
```

```python
pd_res=pd.DataFrame(res)
pd_res.rename(columns={0:'ReimRange',
                       1:'FraudPercentage'}, inplace = True)
```

```python
pd_res.head()
```

```python

ax=sns.jointplot(data=pd_res,x='ReimRange',
                       y='FraudPercentage')
ax.plot(sns.scatterplot, sns.histplot, alpha=.7, edgecolor=".2", linewidth=.5)
ax.plot_marginals(sns.histplot, kde=True)
ax.set_axis_labels(xlabel='ReimbursedRange',ylabel='FraudPercentage')


plt.savefig('./pics/ReimbursedRange&FraudPercentage')#FraudPercentage=Fraud/(NonFraud+Fraud)
```

```python
pd.to_datetime(df_num.ClaimStartDt).dt.month.value_counts()
```

```python
df_num['month']=pd.to_datetime(df_num.ClaimStartDt).dt.month
```

```python
res_month=[]
for i in range(1,13):
    p=df_num['PotentialFraud'][df_num['month']==i].value_counts('yes')[1]
    
    res_month.append([i,p])
```

```python
res_month
```

```python
pd_res_month=pd.DataFrame(res_month)
pd_res_month.rename(columns={0:'Month',
                       1:'FraudPercentage'}, inplace = True)
```

```python
pd_res_month
```

```python
ax=sns.jointplot(data=pd_res_month,x='Month',
                       y='FraudPercentage')
ax.plot(sns.scatterplot, sns.histplot, alpha=.7, edgecolor=".2", linewidth=.5)
ax.plot_marginals(sns.histplot, kde=True)
plt.savefig('./pics/Month&FraudPercentage')
```

```python
df_num['Claim_Period']=(pd.to_datetime(df_num.ClaimEndDt)-pd.to_datetime(df_num.ClaimStartDt)).dt.days+1
```

```python
res_period=[]
for i in range(1,38):
    percent=df_num['PotentialFraud'][df_num['Claim_Period']==i].value_counts(normalize=True,ascending=True)[0]
    
    res_period.append([i,percent])
```

```python
pd_res_period=pd.DataFrame(res_period)
pd_res_period.rename(columns={0:'Claim_Period',
                       1:'FraudPercentage'}, inplace = True)
```

```python
pd_res_period.head()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(data=pd_res_period,x="Claim_Period", y="FraudPercentage")
plt.savefig('./pics/Claim_Period_FraudPercentage')
```
